```python
import torchaudio
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
File = 'Dataset\DeepFake\Female\G3002\G3002_1_S0391_fake.wav'
waveform, sample_rate  = torchaudio.load(File)
resample = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
waveform = resample(waveform)
```

```python
#Taken from: https://pytorch.org/tutorials/beginner/audio_preprocessing_tutorial.html
#This is modified
def plot_specgram(waveform, sample_rate, xlim=None):
    waveform = np.array(waveform)

    num_channels, num_frames = waveform.shape
    time_axis = torch.arange(0, num_frames) / sample_rate

    figure, axes = plt.subplots(figsize=(4, 3)) #790 x 490 pixels
    if num_channels == 1:
        axes = [axes]
        axes[0].specgram(waveform[0], Fs=sample_rate)
        axes[0].axis('off')

    plt.tight_layout()
    plt.show(block=False)
```

```python
plot_specgram(waveform, sample_rate=16000)
```
