```python
from scipy.stats import wilcoxon
import pandas as pd
import numpy as np
from scipy.stats import binom
```

```python
audio_data = pd.read_csv("OpenSMILE_all_features.csv")
columns_to_keep = ['label', 'file','StddevUnvoicedSegmentLength', 'MeanUnvoicedSegmentLength',]
Data = audio_data[columns_to_keep]
Deepfake = Data[Data['label'] ==  0]
Genuine = Data[Data['label'] ==  1]
```

```python
#Both Genuine and Deepfake should match each other
#(I.e G3002_1_S0003 == G3002_1_S0003_fake)
print(Genuine.to_string())
```

```python
print(Deepfake.to_string())
```

```python
#Wilcoxon test on MeanUnvoicedSegmentLength

stat, p = wilcoxon(Genuine['MeanUnvoicedSegmentLength'], Deepfake['MeanUnvoicedSegmentLength'], alternative='two-sided', nan_policy='raise')

print("Wilcoxon Signed-Rank Test on MeanUnvoicedSegmentLength \n Statistic: ", stat, "\n p-value: ", p)
```

```python
#Wilcoxon test on StddevUnvoicedSegmentLength

stat, p = wilcoxon(Genuine['StddevUnvoicedSegmentLength'], Deepfake['StddevUnvoicedSegmentLength'], alternative='two-sided')

print("Wilcoxon Signed-Rank Test on StddevUnvoicedSegmentLength \n Statistic: ", stat, "\n p-value: ", p)
```

```python
A = Genuine['MeanUnvoicedSegmentLength']
B = Deepfake['MeanUnvoicedSegmentLength']

#After - Before
Diff = B.values - A.values
Diff_non_zero = Diff[Diff != 0]

num_positive = np.sum(Diff_non_zero > 0)
num_negative = np.sum(Diff_non_zero < 0)

test_statistic = min(num_positive, num_negative)
n = len(Diff_non_zero)
p_value = 2 * binom.cdf(test_statistic, n, 0.5)  # multiplied by 2 for the two-tailed test

print("Number of positive differences:", num_positive)
print("Number of negative differences:", num_negative)
print("Test statistic:", test_statistic)
print("p-value:", p_value)
```
