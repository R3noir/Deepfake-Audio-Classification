```python
import os
import shutil

# Define paths
deepfake_path = 'Dataset/Deepfake'
genuine_path = 'Dataset/Genuine'

# Function to get sorted list of files in a directory
def get_sorted_files(path):
    files = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))]
    files.sort()
    return files

# Traverse the Deepfake directory
for root, dirs, files in os.walk(deepfake_path):
    for file in files:
        if file.endswith(('.mp3', '.wav')):
            # Get the corresponding genuine file
            relative_path = os.path.relpath(root, deepfake_path)
            genuine_dir = os.path.join(genuine_path, relative_path)
            genuine_files = get_sorted_files(genuine_dir)
            
            # Get the index from the deepfake file name
            index = int(os.path.splitext(file)[0]) - 1
            
            # Get the corresponding genuine file name
            if index < len(genuine_files):
                genuine_file = genuine_files[index]
                genuine_file_name, genuine_file_ext = os.path.splitext(genuine_file)
                new_deepfake_file_name = f"{genuine_file_name}_fake{genuine_file_ext}"
                
                # Rename the deepfake file to match the genuine file name with _fake suffix
                deepfake_file_path = os.path.join(root, file)
                new_deepfake_file_path = os.path.join(root, new_deepfake_file_name)
                shutil.move(deepfake_file_path, new_deepfake_file_path)
                print(f"Renamed {deepfake_file_path} to {new_deepfake_file_path}")
```
